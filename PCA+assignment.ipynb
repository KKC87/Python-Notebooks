```python
import pandas as pd
import numpy as np

wine = pd.read_csv('https://archive.ics.uci.edu/ml/'
                  'machine-learning-databases/wine/wine.data',
                   header=None)
wine.columns = ['Class_label','Alcohol','Malic_acid','Ash',
               'Alcalinity_of_ash','Magnesium','Total_phenols',
               'Flavanoids','Nonflavanoid_phenols','Proanthocyanins','Color_intensity','Hue',
               'OD280/OD315_of_diluted_wines','Proline']
```

```python
wine.head()
wine.shape
wine.Class_label.nunique()
wine.groupby('Class_label').size()
wine.columns
means = np.round(pd.pivot_table(wine,values=['Alcohol','Malic_acid','Ash','Alcalinity_of_ash','Magnesium','Total_phenols',
               'Flavanoids','Nonflavanoid_phenols','Proanthocyanins','Color_intensity','Hue',
               'OD280/OD315_of_diluted_wines','Proline'],index='Class_label',aggfunc=np.mean),2)
means
```

```python
X = wine.drop('Class_label',axis=1)
Y = wine['Class_label']
X.head()
Y.head()
```

```python
from sklearn.preprocessing import scale
X_scaled = scale(X)
X_scaled =pd.DataFrame(X_scaled, columns =X.columns)
X_scaled.head()
X_scaled.describe()
```

```python
from sklearn.decomposition import PCA
pca  = PCA().fit(X_scaled)
```

```python
pca
```

```python
str(pca.explained_variance_ratio_)
```

```python
str(pca.explained_variance_ratio_.sum())
```

```python
pca = PCA(5).fit(X_scaled)
```

```python
str(pca.explained_variance_ratio_.sum())
```

```python
pca.explained_variance_
```

```python
pca_df = pd.DataFrame(pca.components_, columns=X.columns)
pca_df
```

```python
import matplotlib.pyplot as plt
% matplotlib inline

plt.figure(figsize=(10,10))

import seaborn as sns

sns.heatmap(pca_df, cmap ='RdYlGn', annot=True)

plt.show();
```
